### Temario del curso de **Introducción a Python 3**


#### **Clase 1: Introducción a Python**

1. **Introducción a Python (conceptos generales)**  
   Se comenzará con una explicación de los conceptos fundamentales que definen a Python como lenguaje de programación: su naturaleza interpretada, de código abierto, multiparadigma y de alto nivel. Se explorará su historia, aplicaciones comunes, y las razones de su popularidad.
   
2. **El intérprete interactivo y variables**  
   Introducción al uso del intérprete interactivo de Python (REPL), donde los alumnos podrán ejecutar pequeñas instrucciones en tiempo real. Se explicará el concepto de variables, cómo asignarlas y cómo utilizarlas para almacenar datos básicos. Además, se abordarán buenas prácticas en la elección de nombres de variables.


#### **Clase 2: Tipos de datos primitivos y operadores**

3. **Tipos de datos primitivos**  
   Se explicarán los tipos de datos más básicos que ofrece Python, como números enteros (`int`), números de punto flotante (`float`), cadenas de texto (`str`), y valores booleanos (`bool`). Se discutirán las diferencias y cómo identificarlos en Python.

4. **Operadores aritméticos y lógicos**  
   Se presentarán los operadores básicos de Python: operadores aritméticos (+, -, *, /, etc.) para realizar cálculos, y operadores lógicos (and, or, not) para expresar condiciones. Se mostrarán ejemplos prácticos en el contexto de expresiones simples.


#### **Clase 3: Estructuras de control (parte 1)**

5. **Condicionales: `if`, `else`, `elif`**  
   Introducción a las estructuras condicionales. Se enseñará cómo utilizar las declaraciones `if`, `else` y `elif` para ejecutar código basándose en diferentes condiciones.

6. **Comparaciones y operadores relacionales**  
   Se abordarán los operadores relacionales (==, !=, <, >, <=, >=) para comparar valores y cómo funcionan en combinación con las estructuras condicionales.


#### **Clase 4: Estructuras de control (parte 2)**

7. **Bucles `for`**  
   Se introducirá el concepto de bucles utilizando `for` para iterar sobre secuencias (listas, cadenas, etc.), explicando cómo funcionan y cuándo deben utilizarse. Se presentarán ejemplos básicos de iteración.

8. **Bucles `while` y control de flujo**  
   Se explicará el uso del bucle `while` para repetir instrucciones hasta que se cumpla una condición. Además, se verá cómo utilizar las declaraciones `break` y `continue` para controlar el flujo de ejecución dentro de los bucles.


#### **Clase 5: Estructuras de datos (parte 1)**

9. **Listas**  
   Introducción al tipo de dato lista (`list`), una estructura mutable que permite almacenar colecciones de datos. Se cubrirán métodos básicos de manipulación de listas, como agregar, eliminar y modificar elementos.

10. **Tuplas**  
    Se estudiarán las tuplas (`tuple`), una estructura de datos inmutable, explicando su utilidad en situaciones donde no se requiere modificar los datos almacenados.


#### **Clase 6: Estructuras de datos (parte 2)**

11. **Diccionarios**  
    Se introducirá el tipo de datos diccionario (`dict`), una estructura de clave-valor utilizada para almacenar datos no secuenciales. Se mostrarán ejemplos de cómo agregar, eliminar y acceder a elementos de un diccionario.

12. **Conjuntos (sets)**  
    Se explicarán los conjuntos (`set`), una colección no ordenada de elementos únicos, y se enseñará cómo utilizarlos para realizar operaciones de conjunto como uniones e intersecciones.


#### **Clase 7: Funciones (parte 1)**

13. **Introducción a las funciones**  
    Se explicará el concepto de funciones en Python, su sintaxis básica con `def`, cómo definirlas y llamarlas. Se mostrará la utilidad de encapsular bloques de código en funciones reutilizables.

14. **Parámetros y argumentos**  
    Se discutirá cómo pasar información a las funciones utilizando parámetros y argumentos, incluyendo el uso de parámetros con valores predeterminados.


#### **Clase 8: Funciones (parte 2)**

15. **Retorno de valores**  
    Se enseñará cómo utilizar la palabra clave `return` para devolver valores desde una función y cómo manejar ese valor en el programa principal.

16. **Alcance (scope) de las variables**  
    Se abordará el concepto de alcance de variables en Python, diferenciando entre variables locales y globales y cómo se comportan dentro y fuera de las funciones.


#### **Clase 9: Manejo de archivos**

17. **Lectura y escritura de archivos**  
    Introducción a las operaciones básicas de manejo de archivos: abrir, leer, escribir y cerrar archivos en Python. Se verá cómo manejar archivos de texto, así como el uso de `with` para garantizar el cierre correcto de los mismos.

18. **Manejo de excepciones**  
    Se explicará cómo manejar errores y excepciones utilizando `try`, `except`, `else` y `finally`. Esto permitirá a los alumnos anticipar y controlar errores comunes al trabajar con archivos u otras operaciones.


#### **Clase 10: Programación Orientada a Objetos (parte 1)**

19. **Clases y objetos**  
    Introducción a los conceptos básicos de la Programación Orientada a Objetos (OOP). Se explicará cómo definir clases, crear objetos y utilizar métodos y atributos.

20. **Constructores y el método `__init__`**  
    Se enseñará cómo inicializar objetos utilizando el método constructor `__init__` para definir atributos iniciales.


#### **Clase 11: Programación Orientada a Objetos (parte 2)**

21. **Encapsulamiento y visibilidad**  
    Se explicará el concepto de encapsulamiento en OOP, protegiendo atributos de acceso directo mediante el uso de convenciones como guiones bajos y cómo aplicar los métodos `get` y `set` para controlar el acceso a los datos.

22. **Herencia**  
    Introducción al concepto de herencia en OOP, mostrando cómo una clase puede heredar propiedades y métodos de otra, permitiendo reutilización de código y la creación de estructuras jerárquicas.


#### **Clase 12: Proyectos finales y repaso general**

23. **Desarrollo de proyectos simples**  
    Los alumnos trabajarán en pequeños proyectos que combinen los conceptos vistos durante el curso, como el manejo de variables, funciones, estructuras de control, y OOP.

24. **Repaso y cierre**  
    Se hará un repaso general del curso, reforzando los puntos clave y resolviendo dudas finales. Se presentarán recursos adicionales para continuar el aprendizaje de Python.
